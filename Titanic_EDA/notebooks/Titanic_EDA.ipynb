```python
import pandas as pd
import numpy as np
```

```python
t_data=pd.read_csv("tested.csv")
```

```python
t_data.head(5)
```

```python
print(t_data.info())
```

```python
t_data=t_data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
```

```python
t_data
```

```python
t_data.isnull().sum()
```

```python
t_data['Age'].fillna(t_data['Age'].median(),inplace=True)
```

```python
t_data['Fare'].fillna(t_data['Fare'].mean(),inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t_data['Survived'].value_counts()
```

```python
sns.countplot(x='Survived',data=t_data)
plt.title("survival Count")
plt.show()
```

```python
t_data['Pclass'].value_counts()
sns.countplot(x='Pclass',data=t_data)
plt.title("passenger class distribution")
plt.show()
```

```python
sns.countplot(x='Sex',data=t_data)
plt.title("gender distribution")
plt.show()
```

```python
sns.histplot(t_data['Age'], bins=20,kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
t_data['Age'].value_counts(4)
```

```python
sns.countplot(x='Sex',hue='Survived',data=t_data)
plt.title("survived by gender")
plt.show()
```

```python
sns.countplot(x='Pclass',hue='Survived',data=t_data)
plt.title("survived by passenger class")
plt.show()
```

```python
sns.countplot(x='Survived',hue='Age',data=t_data)
plt.title("survival by age distribution")
plt.show()
```

```python
n_data=t_data.select_dtypes(include=[float,int])
corr=n_data.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("correlation Matrix")
plt.show()
```
